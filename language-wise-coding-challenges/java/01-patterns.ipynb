### Square Patterns


---
For n = 4, print the following square pattern

```text
* * * *
* * * *
* * * *
* * * *
```

````java
public class Test {

    public static void printPattern(int n) {
        for (int i = 0; i < n; i++) {
            for (int j = 0; j < n; j++) {
                System.out.print("* ");
            }
            System.out.println();
        }
    }
}

Test.printPattern(4);
````

### Triangle Patterns


---
For n = 5, print the following pattern

```text
*
* *
* * *
* * * *
* * * * *
```

````java
public class Test {

    // [Algorithm-1]
    public static void printPattern1(int rows) {
        for (int i = 0; i < rows; i++) {
            for (int j = 0; j <= i; j++) {
                System.out.print("* ");
            }
            System.out.println();
        }
    }

    // [Algorithm-2]
    public static void printPattern2(int rows) {
        for (int i = 0; i < rows; i++) {
            for (int j = 0; j < rows; j++) {
                if (i >= j) {
                    System.out.print("* ");
                }
            }
            System.out.println();
        }
    }
}

System.out.println("[Algorithm-1]");
Test.printPattern1(5);
System.out.println("\n[Algorithm-2]");
Test.printPattern2(5);
````

```text
1

1 0 1 0 1
```

````java
public class Test {

    // [Algorithm-1]: Simplified version using a couple of extra variables
    public static void printPattern1(int n) {
        for (int i = 0; i < n; i++) {
            int odd = 1, even = 2;

            for (int j = 0; j <= i; j++) {
                if (i % 2 == 0)
                    System.out.print(odd++ % 2 + " ");
                else
                    System.out.print(even++ % 2 + " ");
            }
            System.out.println();
        }
    }

    // [Algorithm-2]: Simplified version using a single extra variable
    public static void printPattern2(int n) {
        for (int i = 0; i < n; i++) {
            int count = 1;

            for (int j = 0; j <= i; j++) {
                if (i % 2 == 0)
                    System.out.print(count++ % 2 + " ");
                else
                    System.out.print(++count % 2 + " ");
            }
            System.out.println();
        }
    }

    // [Algorithm-3]: Without using any extra variable
    public static void printPattern3(int n) {
        for (int i = 0; i < n; i++) {
            for (int j = 0; j <= i; j++) {
                if (i % 2 == 0)
                    System.out.print((j + 1) % 2 + " ");
                else
                    System.out.print(j % 2 + " ");
            }
            System.out.println();
        }
    }
}

System.out.println("[Algorithm-1]");
Test.printPattern1(5);
System.out.println("\n[Algorithm-2]");
Test.printPattern2(5);
System.out.println("\n[Algorithm-3]");
Test.printPattern3(5);
````

For n = 5, print the following pattern

```text
        *
      * * *
    * * * * *
  * * * * * * *
* * * * * * * * *
```

````java
public class Test {

    // [Algorithm-1]: Using an extra variable
    public static void printPattern1(int n) {
        int count = 1;

        for (int i = 1; i <= n; i++) {
            for (int j = 0; j < n * 2 - i * 2; j++) {
                System.out.print(" ");
            }

            for (int j = 0; j < count; j++) {
                System.out.print("* ");
            }
            System.out.println();
            count = count + 2;
        }
    }

    // [Algorithm-2][Variant-1]: Without using any extra variable
    public static void printPattern2v1(int n) {
        for (int i = 0; i < n; i++) {
            for (int j = 0; j < n - (i + 1); j++) {
                System.out.print("  ");
            }

            for (int j = 0; j < 2 * i + 1; j++) {
                System.out.print("* ");
            }
            System.out.println();
        }
    }

    // [Algorithm-2][Variant-2]
    public static void printPattern2v2(int n) {
        for (int i = 1; i <= n; i++) {
            for (int j = 0; j < n - i; j++) {
                System.out.print("  ");
            }

            for (int j = 0; j < 2 * i - 1; j++) {
                System.out.print("* ");
            }
            System.out.println();
        }
    }
}
System.out.println("[Algorithm-1]");
Test.printPattern1(5);
System.out.println("\n[Algorithm-2][Variant-1]");
Test.printPattern2v1(5);
System.out.println("\n[Algorithm-2][Variant-2]");
Test.printPattern2v2(5);
````

```text
        1

  1 2 3 4 3 2 1
1 2 3 4 5 4 3 2 1
```

````java
public class Test {

    public static void printPattern(int n) {
        for (int i = 1; i <= n; i++) {
            int count = 1;

            for (int j = 0; j < n - i; j++) {
                System.out.print("  ");
            }

            for (int j = 0; j < 2 * i - 1; j++) {
                if (j < i - 1)
                    System.out.print(count++ + " ");
                else
                    System.out.print(count-- + " ");
            }
            System.out.println();
        }
    }
}

Test.printPattern(5);
````
